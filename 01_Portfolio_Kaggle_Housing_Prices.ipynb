# **Iniciando a análise exploratória**

## **Verificando Dados Faltantes**
  - Exploração inicial das características do conjunto de treino.
  - Visualização da distribuição dos preços das casas.
  - Identificação e tratamento de dados faltantes.
  - Engenharia de recursos para melhorar a qualidade dos dados.

```python
# Identifique dados faltantes no conjunto de treino
missing_data = train_data.isnull().sum().sort_values(ascending=False)

# Criando um DataFrame com as informações sobre dados faltantes
missing_info = pd.DataFrame({
    'Variável': missing_data.index,
    'Quantidade de Dados Faltantes': missing_data.values,
    'Porcentagem de Dados Faltantes': (missing_data / len(train_data)) * 100
})

# Exiba as primeiras linhas do DataFrame
missing_info.head(len(missing_info))
```

```python
# Identifique dados faltantes no conjunto de treino
missing_data = train_data.isnull().sum().sort_values(ascending=False)
missing_data = missing_data[missing_data > 0]

plt.figure(figsize=(10, 6))
missing_data.plot(kind='bar', color='orange')
plt.title('Dados Faltantes no Conjunto de Treino')
plt.xlabel('Variáveis')
plt.ylabel('Quantidade de Dados Faltantes')
plt.show()
```

# Qual a melhor abordagem para solucionar os dados faltantes?

**Remoção de Colunas ou Linhas**

Se uma coluna tiver uma quantidade significativa de dados faltantes e não for crucial para a análise, posso considerar removê-la.
Se uma linha específica (observação) tiver muitos dados faltantes e não puder ser razoavelmente inferida, considerar a remoção da linha.

**Preenchimento com Estatísticas**

Para variáveis numéricas, preencher os valores faltantes com a média, mediana ou moda da respectiva coluna.
Para variáveis categóricas, preencher com a categoria mais frequente.

**Imputação com Modelagem**

Usar modelos preditivos para estimar valores faltantes com base em outras variáveis.

**Preenchimento Manual**

Pode preencher manualmente alguns valores faltantes, mas não muitos valores faltantes.

## Podemos analizar essas variáveis categóricas uma de cada vez:

**PoolQC:**

Dada a porcentagem significativa de dados faltantes (99.52%), podemos assumir que a maioria das casas não possui piscina. Podemos preencher os valores faltantes com "NoPool" para indicar a ausência de piscina.

**MiscFeature:**

Similar à abordagem do PoolQC, podemos assumir que a maioria das casas não possui um recurso diverso. Podemos preencher os valores faltantes com "NoMiscFeature".

**Alley:**

A variável Alley representa o tipo de beco próximo à propriedade. Dados faltantes aqui podem indicar a ausência de beco. Podemos preencher com "NoAlley".

**Fence:**

A variável Fence representa o tipo de cerca. Dados faltantes podem indicar a ausência de cerca. Podemos preencher com "NoFence".

**MasVnrType:**

Esta variável indica o tipo de revestimento de alvenaria. Vamos preenchê-la com a moda, que é o valor mais frequente.

```python
# Preenchimento específico para cada variável categórica
fill_values = {
    'PoolQC': 'NoPool',
    'MiscFeature': 'NoMiscFeature',
    'Alley': 'NoAlley',
    'Fence': 'NoFence',
    'MasVnrType': train_data['MasVnrType'].mode()[0]
}

train_data.fillna(value=fill_values, inplace=True)
```

### Vamos verificar novamente como está ficando:

```python
# Identifique dados faltantes no conjunto de treino
missing_data = train_data.isnull().sum().sort_values(ascending=False)

# Crie um DataFrame com as informações sobre dados faltantes
missing_info = pd.DataFrame({
    'Variável': missing_data.index,
    'Quantidade de Dados Faltantes': missing_data.values,
    'Porcentagem de Dados Faltantes': (missing_data / len(train_data)) * 100
})

# Exiba as primeiras linhas do DataFrame
missing_info.head(len(missing_info))
```

```python
# Identifique dados faltantes no conjunto de treino
missing_data = train_data.isnull().sum().sort_values(ascending=False)
missing_data = missing_data[missing_data > 0]

plt.figure(figsize=(10, 6))
missing_data.plot(kind='bar', color='orange')
plt.title('Dados Faltantes no Conjunto de Treino')
plt.xlabel('Variáveis')
plt.ylabel('Quantidade de Dados Faltantes')
plt.show()
```

### Agora que a quantidade diminiu, vamos filtrar todo o restante de dados faltantes, após os preenchimentos.

```python
# Identificando dados faltantes no conjunto de treino após o preenchimento
missing_data_after_fill = train_data.isnull().sum().sort_values(ascending=False)
missing_data_after_fill = missing_data_after_fill[missing_data_after_fill > 0]

# Criando um DataFrame com as informações sobre dados faltantes após o preenchimento
missing_info_after_fill = pd.DataFrame({
    'Variável': missing_data_after_fill.index,
    'Quantidade de Dados Faltantes': missing_data_after_fill.values,
    'Porcentagem de Dados Faltantes': (missing_data_after_fill / len(train_data)) * 100
})

# Exiba as primeiras linhas do DataFrame
missing_info_after_fill.head(len(missing_info_after_fill))
```

## Analisando e tomando uma decisão em relação aos últimos dados faltantes

**FireplaceQu:**

Esta variável representa a qualidade da lareira. Como a porcentagem de dados faltantes é relativamente alta (47.26%), podemos assumir que a ausência de dados indica a ausência de lareira. Podemos preencher com "NoFireplace".

**LotFrontage:**

A variável LotFrontage representa a metragem da frente do lote. Podemos preencher os dados faltantes com a média ou mediana dessa variável.

**GarageFinish, GarageType, GarageQual, GarageCond, GarageYrBlt:**

Essas variáveis estão relacionadas à garagem. Podemos assumir que a ausência de dados indica a ausência de garagem. Para GarageYrBlt, podemos preencher com o ano de construção da casa.

**BsmtExposure, BsmtFinType2, BsmtQual, BsmtCond, BsmtFinType1:**

Essas variáveis estão relacionadas ao porão. Podemos assumir que a ausência de dados indica a ausência de porão.

**MasVnrArea:**

Representa a área do revestimento de alvenaria. Podemos preencher os dados faltantes com a média ou mediana dessa variável.

**Electrical:**

Representa o sistema elétrico da casa. Com apenas 1 dado faltante, podemos preenchê-lo com a moda.

```python
# Preenchimento específico para cada variável
fill_values = {
    'FireplaceQu': 'NoFireplace',
    'LotFrontage': train_data['LotFrontage'].median(),
    'GarageFinish': 'NoGarage',
    'GarageType': 'NoGarage',
    'GarageQual': 'NoGarage',
    'GarageCond': 'NoGarage',
    'GarageYrBlt': train_data['YearBuilt'],
    'BsmtExposure': 'NoBsmt',
    'BsmtFinType2': 'NoBsmt',
    'BsmtQual': 'NoBsmt',
    'BsmtCond': 'NoBsmt',
    'BsmtFinType1': 'NoBsmt',
    'MasVnrArea': train_data['MasVnrArea'].median(),
    'Electrical': train_data['Electrical'].mode()[0]
}

train_data.fillna(value=fill_values, inplace=True)
```

### Verificando dados faltantes no conjunto de treino após as últimas imputações

```python
# Identificando dados faltantes no conjunto de treino após o preenchimento
missing_data_after_fill = train_data.isnull().sum().sort_values(ascending=False)
missing_data_after_fill = missing_data_after_fill[missing_data_after_fill > 0]

# Verificar se há dados faltantes após o preenchimento
if missing_data_after_fill.empty:
    print("Não há mais dados faltantes.")
else:
    # Criando um DataFrame com as informações sobre dados faltantes após o preenchimento
    missing_info_after_fill = pd.DataFrame({
        'Variável': missing_data_after_fill.index,
        'Quantidade de Dados Faltantes': missing_data_after_fill.values,
        'Porcentagem de Dados Faltantes': (missing_data_after_fill / len(train_data)) * 100
    })

    # Exiba as primeiras linhas do DataFrame
    missing_info_after_fill.head(len(missing_info_after_fill))
```
